```python
from transformers import Qwen2_5_VLForConditionalGeneration, AutoTokenizer, AutoProcessor
from qwen_vl_utils import process_vision_info
from awq import AutoAWQForCausalLM
import torch
import numpy
import cv2
print(numpy.__version__)
print(torch.__version__, torch.cuda.is_available())  # 显示 PyTorch 版本和 CUDA 可用性
from transformers import Qwen2_5_VLForConditionalGeneration
print('Transformers版本兼容！')  # 确认 Transformers 版本兼容

```

```python
# 加载模型与处理器（自动分配显存）
model_path = "/home/jack/桌面/Qwen2.5-VL-7B-Instruct-AWQ"

model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    model_path,
    device_map={"": "cuda:0"},
    trust_remote_code=True,
    torch_dtype=torch.float16,  # AWQ模型推荐使用半精度
)
processor = AutoProcessor.from_pretrained(model_path)
```

```python
# 视频路径（替换为您的 .mp4 文件路径）
video_path = "/home/jack/桌面/qwen_vl_image/匝道会车.mp4"

# 读取视频并提取帧
def extract_video_frames(video_path, max_frames=8):
    """提取视频帧，最多 max_frames 帧"""
    cap = cv2.VideoCapture(video_path)
    frames = []
    frame_count = 0
    
    while cap.isOpened() and frame_count < max_frames:
        ret, frame = cap.read()
        if not ret:
            break
        # 转换为 PIL 图像
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        pil_image = Image.fromarray(frame)
        frames.append(pil_image)
        frame_count += 1
    
    cap.release()
    return frames

# 提取视频帧
video_frames = extract_video_frames(video_path, max_frames=8)  # 限制帧数以控制显存
```

最终整合视频流信息：

```python
from IPython.display import HTML
import cv2
import base64
import os
video_path = "/home/jack/桌面/qwen_vl_image/匝道会车.mp4"
def convert_video_to_html5(video_path):
    """将视频转换为 HTML5 视频"""
    # 临时输出路径
    temp_output = "temp_output.mp4"
    # 使用 ffmpeg 确保兼容 HTML5
    os.system(f"ffmpeg -i {video_path} -vcodec h264 -acodec aac -y {temp_output}")
    
    # 读取视频并编码为 base64
    with open(temp_output, "rb") as video_file:
        video_data = base64.b64encode(video_file.read()).decode("utf-8")
    
    # 创建 HTML5 视频标签
    video_html = f"""
    <video width="640" height="480" controls>
        <source src="data:video/mp4;base64,{video_data}" type="video/mp4">
        Your browser does not support the video tag.
    </video>
    """
    os.remove(temp_output)  # 删除临时文件
    return HTML(video_html)


# 输入示例（替换为本地图像路径）
messages = [
    {
        "role": "user",
        "content": [
            {"type": "video", "video": video_frames},
            {"type": "text", "text": """分析自动驾驶BEV视频流，图像元素如下：白色矩形：自身车辆（ego vehicle），位于视野中央。
            绿色矩形及其矢量线段：其他车辆及其位移矢量。紫色圆形及其矢量：行人及其位移矢量。橙色曲线：ego车辆已执行的规
            划路径。绿色曲线：ego车辆未来规划未执行的路径。深蓝区域：车辆可行区域。
            任务：1.识别长尾场景（如行人突然横穿、盲区物体、异常车辆行为、哪些时间区段需要等待礼让），避免强制预警，适当预测行人与车辆的移动意图，如果
            现有行人/车辆与自车干涉概率低可适当过滤，仅输出白色车辆高度相关潜在风险的分析;
            2.对当前驾驶风格建议：（保守、正常、激进）
            3.代入20帧场景的驾驶者，提醒视频起始前20帧区间时刻大致方位（左/右侧后方）来车（无车辆时可省去）
            要求：言简意赅：100字左右按重要性顺序高效率汇报"""}
        ]
    }
]

# 预处理与推理
text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
image_inputs, _ = process_vision_info(messages)
inputs = processor(
    text=[text],
    images=image_inputs,
    return_tensors="pt"
).to("cuda")

# 生成响应（限制输出结构）
generated_ids = model.generate(
    **inputs,
    max_new_tokens=120,
    temperature=0.2,  # 降低随机性
    do_sample=False
)
output_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

# 在帧提取前显示视频
display(convert_video_to_html5(video_path))
print("模型输出：", output_text)
```
